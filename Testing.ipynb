```python
from pypdf import PdfReader
```

```python
pdf_file = PdfReader('cvtSAj.pdf')
```

```python
for page in pdf_file.pages:
    print(page.extract_text())
```

```python
content =''
for page in pdf_file.pages:
    content =content + page.extract_text() + '\n'
```
